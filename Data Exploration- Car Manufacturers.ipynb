```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.read_csv('Cars93.csv')
```

### Box Plot to Show Revs per Mile for each Car Manufacturer

```python
df_sub = df.loc[df['Manufacturer'].isin(['Audi', 'Hyundai','Suzuki','Toyota'])]
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=df_sub)
plt.xlabel('Manufacturer')
plt.ylabel('Revs per mile')
plt.xticks(rotation=90)
plt.show()
```

##### Toyota is the manufacturer of the car with the highest revs per mile.

### Histograms to show MPG in the City and the Highway

```python
plt.hist(df['MPG.city'], bins=20, alpha=0.5, label='City')
plt.hist(df['MPG.highway'], bins=20, alpha=0.5, label='Highway')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

##### It is generally more efficient to drive on the highway.

### Lineplot to show relationship between Wheelbase and Turning circle

```python
sns.lineplot(x='Wheelbase',y='Turn.circle', data=df)
plt.ylabel('Turning circle')
plt.show()
```

##### The relationship between the  wheelbase and turning circle is a positive correlation. As the wheelbase gets larger, the turning circle increases.

### Mean Horsepower by Type

```python
type_avg = df.groupby('Type').mean()
sns.catplot(x='Type', y='Horsepower',kind='bar',data=df)
plt.xlabel('Type')
plt.ylabel('Horsepower')
plt.show()
```

##### On average, a larger car has more has more horsepower.

### Engine sizes

```python
counts = df['EngineSize'].value_counts()
plt.pie(counts, labels=counts.index)
plt.legend()
plt.show()
```

A 3.0L car engine is the most popular size, followed by a 2.2L engine.
